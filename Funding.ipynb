```python
import os
startup["yearmonth"] = (pd.to_datetime(startup['Date dd/mm/yyyy'],format='%d/%m/%Y').dt.year*100)+(pd.to_datetime(startup['Date dd/mm/yyyy'],format='%d/%m/%Y').dt.month)
temp = startup['yearmonth'].value_counts().sort_values(ascending = False)
print("Number of funding per month in decreasing order (Funding Wise)\n\n",temp)
year_month = startup['yearmonth'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
sns.barplot(year_month.index, year_month.values, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('Year-Month of transaction', fontsize=12)
plt.ylabel('Number of fundings made', fontsize=12)
plt.title("Year-Month - Number of Funding Distribution", fontsize=16)
plt.show()

```

```python
Investment_Type = startup['InvestmentnType'].value_counts()
print(Investment_Type)
plt.figure(figsize=(12,5))
sns.barplot(Investment_Type.index, Investment_Type.values, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('Investment Type', fontsize=12)
plt.ylabel('Number of fundings made', fontsize=12)
plt.title("Investment Type - Number of fundings distribution", fontsize=16)
plt.show()
```

```python
City_Loc = startup['City  Location'].value_counts()
print(City_Loc)
plt.figure(figsize=(12,5))
sns.barplot(City_Loc.index, City_Loc.values, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('City Name', fontsize=12)
plt.ylabel('No of Fundings', fontsize=12)
plt.title("City Location - Improving Scalabilty", fontsize=16)
plt.show()
```

```python
Industry_Vertical = startup['Industry Vertical'].value_counts()
print(Industry_Vertical)
plt.figure(figsize=(100,50))
sns.barplot(Industry_Vertical.index, Industry_Vertical.values, alpha=0.9)
plt.xticks(rotation='vertical')
plt.xlabel('Sectors', fontsize=12)
plt.ylabel('Fundings raised', fontsize=12)
plt.title("Market Share", fontsize=16)
plt.show()
```
